### **1. What is Java?**

Java is a high-level, object-oriented, class-based programming language developed by Sun Microsystems (now owned by Oracle). It is platform-independent, meaning code written in Java can run on any system that has a Java Virtual Machine (JVM). It is used to build desktop, web, and mobile applications.

---

### **2. What are the main features of Java?**

* **Platform Independent**: "Write once, run anywhere" â€” Java code runs on any device with a JVM.
* **Object-Oriented**: Everything in Java revolves around classes and objects.
* **Simple and Secure**: Java removes complex features like pointers and supports security mechanisms.
* **Robust**: Offers strong memory management and exception handling.
* **Multithreaded**: Allows multiple threads to run simultaneously.
* **High Performance**: Uses a Just-In-Time (JIT) compiler.
* **Distributed**: Supports RMI and networking to build distributed applications.

### **3. What is JVM, JRE, and JDK?**

* **JVM (Java Virtual Machine)**: Executes Java bytecode; ensures platform independence.
* **JRE (Java Runtime Environment)**: Includes JVM + libraries needed to run Java programs.
* **JDK (Java Development Kit)**: Includes JRE + development tools (e.g., compiler, debugger).

### **5. What are the 4 main OOP principles in Java?**

1. **Encapsulation**: Wrapping data and methods together in a class; restrict access using access modifiers.
2. **Abstraction**: Hiding implementation details using abstract classes or interfaces.
3. **Inheritance**: One class acquires properties and methods of another.
4. **Polymorphism**: One interface, multiple implementations (method overloading and overriding).

### **10. What is method overriding?**

Method overriding occurs when a subclass provides its own implementation of a method in its superclass.

**Example**:

```java
class Animal {
    void sound() {
        System.out.println("Animal sound");
    }
}
class Dog extends Animal {
    @Override
    void sound() {
        System.out.println("Dog barks");
    }
}
```

### **19. Can a constructor be private? Why?**

Yes, a constructor can be **private**. It is commonly used in the **Singleton Design Pattern** to restrict the creation of objects from outside the class.

**Use Case**: To control instance creation and ensure only one object is created.

### **42. What is the use of try, catch, finally, throw, and throws?**

* **try**: Defines a block where exceptions may occur.
* **catch**: Catches and handles exceptions from the try block.
* **finally**: Executes code regardless of exception (cleanup code).
* **throw**: Used to explicitly throw an exception.
* **throws**: Declares exceptions that a method might throw.

**Example:**

```java
try {
   int x = 1 / 0;
} catch (ArithmeticException e) {
   System.out.println("Error: " + e);
} finally {
   System.out.println("Always runs");
}
```

### **43. Can we write finally block without catch?**

Yes, we can use a `finally` block without a `catch`, but it must be used with a `try`. It ensures that cleanup code always runs.

**Example:**

```java
try {
   // risky code
} finally {
   // cleanup code
}
```

### **60. What is multithreading? Basic syntax?**

**Multithreading** allows concurrent execution of multiple threads for better performance.

**Two ways to create threads:**

**1. Extend Thread class:**

```java
class MyThread extends Thread {
    public void run() {
        System.out.println("Running thread");
    }
}
new MyThread().start();
```

**2. Implement Runnable:**

```java
class MyRunnable implements Runnable {
    public void run() {
        System.out.println("Running thread");
    }
}
new Thread(new MyRunnable()).start();
```
